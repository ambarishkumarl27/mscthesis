# Quantum k-NN Implementation

```python
import numpy as np
import pandas as pd
from qiskit.circuit.library import ZZFeatureMap  # Quantum feature map
from qiskit_aer import AerSimulator  # Backend for simulation
from qiskit_machine_learning.kernels import FidelityQuantumKernel  # Corrected quantum kernel
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Replace this with your actual dataset
data = {
    'ND': np.random.rand(200),  # National Demand values
    'TSD': np.random.rand(200)   # Transmission System Demand values
}
df = pd.DataFrame(data)

# Prepare training data (First 100 rows)
X_train = df[['ND', 'TSD']].values[:100]  
y_train = (df['ND'] > df['ND'].median())[:100]  # Binary classification based on median National Demand

# Prepare testing data (Next 100 rows)
X_test = df[['ND', 'TSD']].values[100:200]  
y_test = (df['ND'] > df['ND'].median())[100:200]  # Binary classification for the test set

#Define the Fidelity Quantum Kernel
feature_map = ZZFeatureMap(feature_dimension=X_train.shape[1], reps=2)  # Define feature map for kernel
quantum_kernel = FidelityQuantumKernel(feature_map=feature_map)  # Use FidelityQuantumKernel

#Compute Pairwise Kernel Distances (Quantum Distances)
def compute_kernel_matrix(X_train, X_test, quantum_kernel):
    # Compute kernel matrix for all pairs of training and testing samples
    kernel_matrix = quantum_kernel.evaluate(X_test, X_train)
    return kernel_matrix

#Implement Custom Quantum k-NN
def quantum_knn(X_train, y_train, X_test, k=3, quantum_kernel=quantum_kernel):
    kernel_matrix = compute_kernel_matrix(X_train, X_test, quantum_kernel)
    
    predictions = []
    
    for i in range(len(X_test)):
        # For each test point, find the k nearest neighbors based on kernel similarity
        distances = kernel_matrix[i]
        
        # Get the indices of the k nearest neighbors
        nearest_neighbors = np.argsort(distances)[-k:]
        
        # Perform majority voting on the labels of the nearest neighbors
        neighbor_labels = y_train[nearest_neighbors]
        predicted_label = np.bincount(neighbor_labels).argmax()
        
        predictions.append(predicted_label)
    
    return np.array(predictions)

#Train and Evaluate the Quantum k-NN Model
k = 3  # Number of nearest neighbors
predictions = quantum_knn(X_train, y_train, X_test, k=k, quantum_kernel=quantum_kernel)

# Evaluate the model accuracy
accuracy = accuracy_score(y_test, predictions)
print(f"Quantum k-NN Model Accuracy: {accuracy * 100:.2f}%")

#Visualizing Results (Optional)
plt.figure(figsize=(10, 6))
plt.plot(df['ND'][100:200], label='True Values', marker='o')
plt.plot(df['ND'][100:200], predictions, label='Predicted Values', linestyle='--', marker='x')
plt.title('Quantum k-NN Model: True vs Predicted National Demand')
plt.xlabel('Sample Index')
plt.ylabel('National Demand (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
